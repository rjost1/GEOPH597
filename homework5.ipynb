```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
p4_df = pd.read_csv('/Users/robertjost/Spring_2026/geoph597/problem4.csv')
p5_df = pd.read_csv('/Users/robertjost/Spring_2026/geoph597/problem5.csv')
```

# Problem 4

```python
fig, ax = plt.subplots(figsize=(12,4))
ax.plot(p4_df['Horizontal_Position_m'], p4_df['Apparent_resistivity_ohm_m'], marker='s',color='b', label='constant spread')
ax.set_xlabel('Horizontal position (m)')
ax.set_ylabel("Apparent Resistitivty")
ax.legend(loc='best')
plt.grid(True)
plt.show()
```

The squiggle between 25 and 30 is likely a near vertical contact between two layers with different resistivity. Based on Resistivity values in the table on page 325 the substrate on the left side of the contact could be partially saturated silty-clay since the apparent resistivity is around 20. The stuff on the right may be coarser grained unconsolidated material like sand and gravel or regolith.

# Problem 5

```python
fig, ax = plt.subplots(figsize=(12,4))
ax.plot(p5_df['Electrode_Spacing_m'], p5_df['Apparent_resistivity_ohm_m'], marker='s',color='b', label='expanding spread')
ax.set_xlabel('Horizontal position (m)')
ax.set_ylabel("Apparent Resistitivty")
ax.legend(loc='best')
plt.grid(True)
plt.show()
```

The traverse hits the water table around 7m horizontal when apparent resistivity values start to decrease, and the most change is occuring betweeon 7 & 15m. I think the depth to the water table is on the order of node spacing around the peak value. Approximately 7 m.
